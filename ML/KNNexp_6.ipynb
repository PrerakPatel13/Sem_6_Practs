```python
import numpy as np
import pandas as pd
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((np.array(point1) - np.array(point2))**2))

def k_nearest_neighbors_regression(input_age, input_height, k, age, height, weight):
    distances = []
    for i in range(len(age)):
        dist = euclidean_distance([input_age, input_height], [age[i], height[i]])
        distances.append((dist, weight[i]))

    distances.sort()
    k_nearest = distances[:k]
    total_weight = sum(weight for dist, weight in k_nearest)

    predicted_weight = total_weight / k
    return predicted_weight

def k_nearest_neighbors_classification(input_height, input_weight, k, height, weight, classes):
    distances = []
    for i in range(len(height)):
        dist = euclidean_distance([input_height, input_weight], [height[i], weight[i]])
        distances.append((dist, classes[i]))

    distances.sort()
    k_nearest = distances[:k]

    class_count = {}
    for class_label in classes:
        class_count[class_label] = 0

    for _, class_label in k_nearest:
        class_count[class_label] += 1

    predicted_class = max(class_count, key=class_count.get)
    return predicted_class

Age = [45, 26, 30, 34, 40, 36, 19, 28, 23, 32]
Height = [5, 5.11, 5.6, 5.9, 4.8, 5.8, 5.3, 5.8, 5.5, 5.6]
Weight = [77, 47, 55, 59, 72, 60, 40, 60, 45, 58]
input_age = 38
input_height= 5.5
k1=1
predicted_weight = k_nearest_neighbors_regression(input_age, input_height, k1, Age, Height, Weight)
print(f"DATASET 1 : Predicted weight for k={k1}, age for  {input_age}, height {input_height}: {predicted_weight:.2f}")
k2=3
predicted_weight = k_nearest_neighbors_regression(input_age, input_height, k2, Age, Height, Weight)
print(f"DATASET 1 : Predicted weight for k={k2}, age for {input_age}, height {input_height}: {predicted_weight:.2f}")
k3=5
predicted_weight = k_nearest_neighbors_regression(input_age, input_height, k3, Age, Height, Weight)
print(f"DATASET 1 : Predicted weight for k={k3}, age for {input_age}, height {input_height}: {predicted_weight:.2f}")


height = [167, 182, 176, 173, 172, 174, 169, 173, 170]
weight = [51, 62, 69, 64, 65, 56, 58, 57, 55]
classification = np.array(['underweight', 'normal', 'normal', 'normal', 'normal', 'underweight', 'normal', 'normal', 'normal'])
input_height = 170
input_weight = 57
k4=1
predicted_class = k_nearest_neighbors_classification(input_height, input_weight, k4, height, weight, classification)
print(f"DATASET 2 : Predicted class for k={k4}, for Height {input_height}, Weight {input_weight}: {predicted_class}")
k5=3
predicted_class = k_nearest_neighbors_classification(input_height, input_weight, k5, height, weight, classification)
print(f"DATASET 2 : Predicted class for k={k5}, for Height {input_height}, Weight {input_weight}: {predicted_class}")
k6=5
predicted_class = k_nearest_neighbors_classification(input_height, input_weight, k6, height, weight, classification)
print(f"DATASET 2 : Predicted class for k={k6}, for Height {input_height}, Weight {input_weight}: {predicted_class}")
```
